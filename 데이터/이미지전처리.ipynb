```
import glob
path = '/content/gdrive/Shared drives/16통계/data/'
yes24images = glob.glob(path + 'yes24Img/*.jpg')
kyoboimages = glob.glob(path + 'kyoboimg/*.jpg')
bookimages = yes24images + kyoboimages
#len(bookimages)
```

```
isbnjpg = []
for i in range(0,len(bookimages)):
  bookroute = bookimages[i]
  isbn = bookroute.split('/')[7]
  isbnjpg.append(isbn)
#len(isbnjpg)
```

```
isbnjpg1 = list(set(isbnjpg))
#len(isbnjpg1)
```

```
totalbook = []
for i in isbnjpg1:
  for j in range(0, len(bookimages)):
    bookroute = bookimages[j]
    isbn = bookroute.split('/')[7]
    if i==isbn:
      totalbook.append(bookimages[j])
      break
#len(totalbook)
```

```
import shutil

for i in totalbook:
    shutil.move(i, '/content/gdrive/Shared drives/16통계/data/totalbookimg')   
```

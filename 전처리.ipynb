```
from google.colab import drive
import os
import pandas as pd
import numpy as np

drive.mount('/content/gdrive/')
```

```
path = '/content/gdrive/Shared drives/16통계/data/'

yes24book = pd.read_csv(path+"bookSumm.csv")
kyobobook = pd.read_csv(path+"kyoboBooksSumm.csv")
```

```
yes24book.head()
```

```
kyobobook.head()
```

```
yes24book = yes24book[['title','topic','author','isbn', 'info', 'review', 'summary']]
kyobobook = kyobobook[['title','topic','author','isbn', 'info', 'review', 'summary']]
```

```
yes24book.head()
```

```
kyobobook.head()
```

```
yes24author = yes24book['author'] 
for i in range(0,len(yes24author)):
  if '/' in yes24author[i]:
     yes24author[i] = yes24author[i].split('/')[0]
  if '정보 더 보기' in yes24author[i]:
    yes24author[i] =  yes24author[i].split('정보 더 보기')[0]
  else:
    yes24author[i] = yes24author[i]
yes24book['author'] = yes24author
```

```
yes24book['author']
```

```
book = pd.concat([yes24book, kyobobook])
print(len(yes24book))
print(len(kyobobook))
print(len(book))
bookcomb = book.drop_duplicates('isbn', keep='first')
print(len(bookcomb))
```

```
bookcomb['topic'].unique()
```

```
bookcomb.loc[bookcomb['topic'] == '사회 정치', 'topic'] = '정치/사회'
bookcomb.loc[bookcomb['topic'] == '경제 경영', 'topic'] = '경제/경영'
```

```
bookcomb['topic'].unique()
```

```
bookcomb.to_csv(path+'bookcomb.csv',index=False)
```
